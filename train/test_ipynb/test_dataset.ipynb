```python
import os
import glob
import torch
import matplotlib.pyplot as plt
from types import SimpleNamespace
from torch.utils.data import DataLoader
from train import CLIP_Clean_Train
# from dci_eval import build_test_loader  # giả sử bạn tách phần khởi tạo DataLoader vào đây
from tqdm import tqdm
import json
import pandas as pd
from PIL import Image
import clip
import torch
from torch.utils.data import Dataset, DataLoader
from types import SimpleNamespace
from tqdm import tqdm
from train import CLIP_Clean_Train
from dci import JsonDCIDataset
from docci import DocciDataset

class Flickr30kPairDataset(Dataset):
    """
    Mỗi caption thành một sample riêng, và chỉ lấy split bạn muốn (train/val/test).
    Trả về: image_tensor, caption_long, caption_short, img_path, split
    """
    def __init__(
        self,
        csv_path: str,
        root_dir: str,
        clip_model: str = "ViT-B/16",
        split: str = None,
        max_items: int = None,
    ):
        super().__init__()
        df = pd.read_csv(csv_path)
        if split is not None:
            df = df[df["split"] == split]
        if max_items is not None:
            df = df.iloc[:max_items]
        df = df.reset_index(drop=True)

        samples = []
        for _, row in df.iterrows():
            img_path = os.path.join(root_dir, "flickr30k-images", row["filename"])
            caps = json.loads(row["raw"])
            # for cap in caps:
            samples.append({
                "img_path": img_path,
                "caption": caps[0].strip(),
                "split": row["split"],
                "img_id": row["img_id"],
            })
        self.samples = samples

        _, self.preprocess = clip.load(clip_model, device="cpu")

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        item = self.samples[idx]
        image = Image.open(item["img_path"]).convert("RGB")
        image_tensor = self.preprocess(image)
        caption_long = item["caption"]
        caption_short = item["caption"]
        return image_tensor, caption_long, caption_short, item["img_path"], item["split"]

def find_ckpt_for_epoch(folder: str, epoch: int):
    # tìm file có cuối là "-{epoch}.pt"
    pattern = os.path.join(folder, f"*-{epoch}.pt")
    matches = glob.glob(pattern)
    if not matches:
        raise FileNotFoundError(f"No checkpoint for epoch {epoch} in {folder}")
    # nếu có nhiều match, lấy file đầu tiên
    return matches[0]

def eval_folder(folder: str, trainer: CLIP_Clean_Train, test_loader: DataLoader):
    ft2i = []
    i2ft = []
    # chạy từ epoch 0 đến 9
    for epoch in range(10):
        ckpt_path = find_ckpt_for_epoch(folder, epoch)
        print(f"Evaluating {ckpt_path} ...")
        state = torch.load(ckpt_path, map_location='cpu')
        trainer.model.load_state_dict(state, strict=False)
        # giả sử test_epoch_ver4 trả về [ft2i, i2ft, ...] dưới dạng decimals (0–1)
        metrics = trainer.test_epoch_ver4(test_loader, epoch=epoch)
        ft2i.append(metrics[0] * 100)
        i2ft.append(metrics[1] * 100)
    return ft2i, i2ft

if __name__ == "__main__":
    # --- Cấu hình chung ---
    os.environ["CUDA_VISIBLE_DEVICES"] = "1"
    args = SimpleNamespace(
        base_model='ViT-B/16',
        download_root=None,
        log_scale=4.6052,
        batch_size=64,
        epochs=1,
        lr=1e-5,
        refine_lr_image=2e-6,
        refine_lr_text=2e-4,
        weight_decay=0.01,
        warmup_length=200,
        exp_name='eval_run'
    )
    # khởi trainer
    trainer = CLIP_Clean_Train(args)

    # khởi test_loader 1 lần
    # --- Khởi tạo dataset + loader ---
    csv_file = "/home/ubuntu/shared/hieu.tq/flickr30k/flickr_annotations_30k.csv"
    root_dir = "/home/ubuntu/shared/hieu.tq/flickr30k"
    test_ds = Flickr30kPairDataset(
        csv_path=csv_file,
        root_dir=root_dir,
        clip_model="ViT-B/16",
        split="test",
        max_items=None
    )
    # test_ds = DocciDataset(split='test', max_items=None)

    test_loader = DataLoader(
        test_ds,
        batch_size   = args.batch_size,
        shuffle      = False,
        num_workers  = 32,     # bây giờ spawn sẽ cho phép dùng CUDA safe
        pin_memory   = True,
    )

    # hai folder tương ứng hai setting
    folder_a = "/home/ubuntu/shared/chau.thv/AAAI_weight/temp_weight/mps_only"
    folder_b = "/home/ubuntu/shared/chau.thv/AAAI_weight/temp_weight/mps_only_img"

    # chạy eval
    ft2i_a, i2ft_a = eval_folder(folder_a, trainer, test_loader)
    ft2i_b, i2ft_b = eval_folder(folder_b, trainer, test_loader)

    # vẽ biểu đồ
    epochs = list(range(10))

    # Full Text → Image
    plt.figure(figsize=(8,4))
    plt.plot(epochs, ft2i_a, marker='o', label='only_mps_learnable')
    plt.plot(epochs, ft2i_b, marker='s', label='only_mps_img_learnable')
    plt.title('Full Text → Image Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy (%)')
    plt.ylim(60, 80)
    plt.xticks(epochs)
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.legend()
    plt.tight_layout()
    plt.show()

    # Image → Full Text
    plt.figure(figsize=(8,4))
    plt.plot(epochs, i2ft_a, marker='o', label='only_mps_learnable')
    plt.plot(epochs, i2ft_b, marker='s', label='not_learnable_weight')
    plt.title('Image → Full Text Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy (%)')
    plt.ylim(60, 80)
    plt.xticks(epochs)
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.legend()
    plt.tight_layout()
    plt.show()
```
